### Gas prices in some countries

```python
gas = pd.read_csv('gas_prices.csv')
gas
```

```python
#plotting some countries against each other

plt.figure(figsize=(8,5), dpi=100)

plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
plt.plot(gas.Year, gas.UK, 'b.-', label='UK')
plt.plot(gas.Year, gas['South Korea'], 'y.-', label='South Korea')
plt.plot(gas.Year, gas.USA, 'g.-', label='US')
plt.title('Gas prices across the world', fontsize=18, fontname='Calibri', fontweight='bold')
plt.xlabel('Years', fontsize=13)
plt.ylabel('Price in US Dollars', fontsize=13)
plt.xticks(gas.Year[::2])

plt.legend()
plt.show()
```

### Calculating and plotting the squad worth of 7 top clubs (Barca, Madrid, Juventus, Utd, PSG, City, Liverpool)

```python
#we begin by loading select columns only since the file is too large

fifa18 = pd.read_csv('fifa_data.csv', usecols=[1,2,7,9,11,12])
fifa18
```

```python
#lets split the dataframe into two to all the conversion of some of the K values to M

kval = fifa18.loc[fifa18['Value'].str.contains('K')] #Values in thousands,K
mval = fifa18.loc[fifa18['Value'].str.contains('M')] #Values in millions, M
noval = fifa18.loc[~fifa18['Value'].str.contains('M|K')] #Values neither in M or K
```

```python
#We begin by cleaning the Player value data of 'kval' and striping it of strings and converting to millions, M

kval.Value = [ (float(x.strip('K'))/1000) for x in kval.Value]
kval.Value
```

```python
#cleaning the Player value data of 'mval' and striping it of strings

mval.Value = [ float(x.strip('M')) for x in mval.Value]
mval.Value
```

```python
#NOTE: We dont have to bother about the noval values since they're valueless (0 values). Their sum = 0

noval.Value = [(float(x.strip(''))) for x in noval.Value]
noval.Value.sum(axis=0)
```

```python
#let's combine the clean data 

fifa18pro = pd.concat([kval,mval])
fifa18pro
```

```python
#lets prepare data for each team

#FC Barcelona
barca = fifa18pro.loc[fifa18pro['Club'] == 'FC Barcelona']
barcelona = barca['Value'].sum(axis=0)

#Real Madrid
madrid = fifa18pro.loc[fifa18pro['Club'] == 'Real Madrid']
r_madrid = madrid['Value'].sum(axis=0)

#Juventus
juve = fifa18pro.loc[fifa18pro['Club'] == 'Juventus']
juventus = juve['Value'].sum(axis=0)

#Manchester United
mutd = fifa18pro.loc[fifa18pro['Club'] == 'Manchester United']
man_utd = mutd['Value'].sum(axis=0)

#Manchester City
mcty = fifa18pro.loc[fifa18pro['Club'] == 'Manchester City']
man_city = mcty['Value'].sum(axis=0)

#Paris Saint-Germain
psg = fifa18pro.loc[fifa18pro['Club'] == 'Paris Saint-Germain']
PSG = psg['Value'].sum(axis=0)

#Liverpool
liv = fifa18pro.loc[fifa18pro['Club'] == 'Liverpool']
liverpool = liv['Value'].sum(axis=0)
```

```python
#plotting the bar chart

fig = plt.figure(figsize=(8,6), dpi=100)

labels = ['FC Barca', 'Real Madrid', 'Juventus', 'Man United', 'Man City', 'PSG', 'Liverpool']
values = [barcelona, r_madrid, juventus, man_utd, man_city, PSG, liverpool]
index = [i for i,_ in enumerate(labels)]

plt.bar(index, values, tick_label=labels)
plt.title('Squad worth for Top 7 European Clubs in FIFA 18', fontweight='bold', fontsize=18)
plt.xlabel('European Clubs', fontsize=14)
plt.ylabel('Net Worth in Million Euros', fontsize=14)
plt.yticks(np.arange(100,1000,100))
plt.savefig('Squad net worth FIFA18.png', dpi=300)

plt.show()
```
