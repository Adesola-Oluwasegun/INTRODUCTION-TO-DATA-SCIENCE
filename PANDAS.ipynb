## Making more changes to data

```python
#Creating a new column: Let's create 'Total column' which is the addition of HP, Attack, Defense, Sp. Atk, Sp. Def, Speed

df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']

df
```

```python
#Or use a shorter method

df['Total2'] = df.iloc[:, 4:10].sum(axis=1) # : for all rows, 4:10 for columns of index 4 to index 9, axis=0 for adding vertically, axis=1 for horizontal operations
df
```

```python
#Dropping a column: lets remove column 'Total2'

df = df.drop(columns=['Total2'])
df
```

```python
#Rearranging columns

#create a list of columns
col = list(df.columns)

#arrange using indexes
df = df[col[0:4] + [col[-1]] + col[4:12]]
df.head(5)
```

## Filtering Our Data

```python
#To set Type 1 to grass and Type 2 to poison: note in pandas 'and' = &; 'or' = |

df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
```

```python
#or 
df[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
```

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['Speed'] > 40)]
```

```python
#using or

df.loc[(df['Type 1'] == 'Grass') | (df['Type 2'] == 'Poison')]
```

```python
#sometimes we might need to work with the new df we've created hence we can rest index

new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['Speed'] > 40)]
new_df = new_df.reset_index()
new_df
```

```python
#from above the df automatically creates a new column for the previous index, we can drop that by using:

new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['Speed'] > 40)]
new_df = new_df.reset_index(drop=True)
new_df
```

```python
#we could just set it inplace rather than equate it to a new df again

new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['Speed'] > 40)]
new_df.reset_index(drop=True, inplace=True)
new_df
```

```python
# we can sort through any column for any values using .contains method: the code below shows all Name that contains 'Mega' in it

df.loc[df['Name'].str.contains('Mega')]
```

```python
# to show those without 'Mega' we add ~ to the code

df.loc[~df['Name'].str.contains('Mega')]
```

```python
#to show which Type 1 contains Psychic or Rock

df.loc[df['Type 1'].str.contains('Psychic|Rock')]
```

```python
# We can use the RegEx module with the contains method

#import re module
import re

df.loc[df['Type 1'].str.contains('psychic|rock', flags=re.I, regex=True)] 

#note that the column names 'psychic and rock' are not of the title case, thats why flags=re.I was introduced:which means ignore case
```

```python
#we are not resttricted to just whole names only. We can use re to search for more. NOTE: BRUSH UP ON REGULAR EXPRESSIONS
#this code shows data for Names containing 'pi' in it; [a-z]=xters btw a &z; *= zero or more

df.loc[df['Name'].str.contains('pi[a-z]*', flags=re.I, regex=True)]
```

```python
# for Names starting with 'Pi' we add ^

df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```

```python
# for Names starting with p and ending with o

df.loc[df['Name'].str.contains('^p.*o$', flags=re.I, regex=True)]
```

```python
#for names ending with pi

df.loc[df['Name'].str.contains('pi$', flags=re.I, regex=True)]
```

## Conditional changes

```python
#we might want to make specific changes to our data: for instance- change all the fire in Type 1 to flames

df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flames'
df
```

```python
#let's change this back

df.loc[df['Type 1'] == 'Flames', 'Type 1'] = 'Fire'
```

```python
#you can manipulate in all sort of ways

df.loc[df['Type 1'] == 'Fire', 'Legendary'] = 'Flames']
df
```

```python
#for all total greater than 600, genr and legend = 'flames'

df.loc[df['Total'] > 600, ['Generation', 'Legendary']] = 'Flames' #or  = ['Flames', 'Fiery'
df
```

## Aggregate Statistics (Groupby)

```python
#lets use our modified file since we've messed the original up. Groupby can give us aggregate stats details like mean, sum, count
#to see the mean of all Type 1 data

df = pd.read_csv('updated pokemon.csv')

df.groupby(['Type 1']).mean()
```

```python
#you can sort

df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
```

```python
#you can sum

df.groupby(['Type 1']).sum()
```

```python
#you can count too: for instance to count the number of values in Type 1, we can create a column count and use it

df['count'] = 1

df.groupby(['Type 1']).count()['count']
```

```python
#you can further count by relating to another column

df.groupby(['Type 1', 'Type 2']).count()['count']
```

## Working with large amounts of data

```python
#We can load heavy files in to data frames in chunks using the chunksize args
chunk = 1
for df in pd.read_csv('updated pokemon.csv', chunksize=100):
    print('\nchunk', chunk)
    print(df)
    chunk += 1
    
```
