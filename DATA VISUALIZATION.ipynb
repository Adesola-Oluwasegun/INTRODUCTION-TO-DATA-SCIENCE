```python
#Customization of plots: changing colors, linestyles, labels, legends and marker types, markersize, markerfacecolor

from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010] 
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# create a line chart, years on x-axis, gdp on y-axis 
plt.plot(years, gdp, color='pink', label='Nom. GDP', marker='>', markerfacecolor='yellow', markersize=10, linestyle='dashed')

# add a title 
plt.title("Nominal GDP")

# add a label to the y-axis 
plt.ylabel("Billions of $") 

#show a legend on the plot
plt.legend()

plt.show()
```

```python
#Line Charts can also be used to show trends, as seen below.

variance     = [1, 2, 4, 8, 16, 32, 64, 128, 256] 
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x+y for x, y in zip(variance, bias_squared)]
index = [ i for i,_ in enumerate(variance)]

#we can make multiple calls to plt.plot to show multiple series on the same chart
plt.plot(index, variance, 'g-', label='variance')# green solid line 
plt.plot(index, bias_squared, 'r-.', label='bias^2')# red dot-dashed line 
plt.plot(index, total_error, 'b:', label='total error')# blue dotted line

#loc=9 means 'top center'
plt.legend(loc=9) 
plt.xlabel("model complexity") 
plt.title("The Bias-Variance Tradeoff") 
plt.show()
```

## When to use scatter graphs: 

Use scatter to show the relationship between two variables. They are sometimes called correlation plots because they show how two variables are correlated. To help visualize the trend it can be useful to draw a trend line along with the scatter graph.

```python
# For example:

import numpy as np 
import matplotlib.pyplot as plt

x = (5, 5.5, 6, 6.5, 7, 8, 9, 10) 
y = (120, 115, 100, 112, 80, 85, 69, 65)

plt.scatter(x,y)

#Generate the trend line
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), 'r')
plt.show()


# The polyﬁt() function performs a least squares polynomial ﬁt for the data it is given.
# A poly1d class is then created based on the array returned by polyﬁt().This class is a one-dimensional polynomial class. 
# It is a convenience class, used to encapsulate “natural” operations on polynomials. 
# The poly1d object is then used to generate a set of values for use with the set of x values for the function pyplot.plot().
```

# PIE CHARTS

A pie chart is a type of graph in which a circle is divided into sectors (or wedges) that each represent a proportion of the whole. 
- Here we will plot a pie chart using plt.pie() method.
- Define the labels
- A portion of each label can be defined using another list (in this case sizes).
- Colors for each label is defined using another list (in this case colors).
- shadow=True will show a shadow beneath each label in pie-chart
- startangle rotates the start of the pie chart by given degrees counterclockwise from the x-axis
- explode is used to set the fraction of radius with which we offset each wedge
- autopct is used to format the value of each label. Here we set it to show the percentage value only upto 1 decimal place.

```python
import matplotlib.pyplot as plt

#defining labels
activities = ['Sex', 'Eat', 'Sleep', 'Pray']

#portion covered by each label
sizes = [25, 20, 25, 30]

#colors for each label
colors = ['g', 'r', 'y', 'b']

#plotting the pie chart
plt.pie(sizes, labels = activities, radius= 2.0, colors = colors, startangle=30, shadow=True, explode = (0.15, 0, 0, 0), autopct = '%1.2f%%')

#plottin legend
plt.legend()

plt.show()
```

#  *BAR CHARTS*

A bar chat is a good choice when we want to show how some quantity varies among some discrete set of items. Because it is easy to interpret bar charts, and how each category relates to another, they are one of the most commonly used types of chart. There are also several different common variations such as grouped bar charts and stacked bar charts. 

PLEASE NOTE: The default WIDTH of each bar = 0.8 with spaces added between each discrete item to difeerentiate them. If you set the  bar width to more than 0.8, the spaces are covered. You'll see examples under grouped bars. 

```python
#Example: shows the previous example on pie charts in bar charts

faculty = ['Met and Mats', 'Mechanical', 'Chemical', 'Systems']
students = [62, 104, 125, 56]
index = (1, 2, 3, 4) #provides locations on x axis

#set up bar chart
plt.bar(index, students, tick_label=faculty)

#configure the layout
plt.ylabel('Number of students', fontsize=12)
plt.xlabel('Departments', fontsize=12)
plt.title('Unilag Engineering students', fontsize=14)

plt.show()
```

### Stacked Bar Charts 

Bar Charts can also be stacked. This can be a way of showing total values(and what contributes to those total values). One thing to note from thhe example below is that after the ﬁrst set of values are added using the pyplot.bar() function, it is necessary to specify the bottom locations for the next set of bars using the bottom parameter. We can do this just using the values already used for web_usage for the second bar chart; however for the third bar chart we must add the values used for web_usage and data_science_usage together (in this case using a for list comprehension).


```python
#Example: 

#set up the data
labels = ('Python', 'Scala', 'C#', 'Java', 'PHP') 
index = [i for i,_ in enumerate(labels)] 
web_usage = [20, 2, 5, 10, 14] 
data_science_usage = [15, 8, 5, 15, 2] 
games_usage = [10, 1, 5, 5, 4]

#set up the bar chart
plt.bar(index, web_usage, tick_label=labels, label='web')
plt.bar(index, data_science_usage, label='data', tick_label=labels, bottom=web_usage)

web_and_games_usage = [web_usage[i] + data_science_usage[i] for i in range(0, len(web_usage))]
plt.bar(index, games_usage, tick_label=labels, label='games', bottom=web_and_games_usage)

#configure the layout
plt.ylabel('Usage')
plt.xlabel('Programming Languages')
plt.legend()
plt.show()
```

```python
#Example 2:

social_media = ['Facebook', 'Twitter', 'Snapchat', 'Instagram']
africans = [700, 300, 100, 500]
americans = [800, 200, 400, 800]
asians = [400, 500, 600, 600]
australians = [300, 300, 250, 200]
index = [i for i,_ in enumerate(social_media)]

#setting up the bar chart
plt.bar(index, africans, tick_label=social_media, label='Africans')
plt.bar(index, americans, tick_label=social_media, label='Americans', bottom=africans)

#setting the bottom for asians
africans_and_americans = [africans[i] + americans[i] for i in range(0, len(index))]
plt.bar(index, asians, tick_label=social_media, label='Asians', bottom=africans_and_americans)

#setting the bottom for australians
aus_bottom = [africans_and_americans[i] + asians[i] for i in range(0, len(index))]
plt.bar(index, australians, tick_label=social_media, label='Australians', bottom=aus_bottom)

plt.title('Social Media Usage Around The World')
plt.xlabel('Social Media Apps')
plt.ylabel('Population in millions')
plt.legend(loc=9)
plt.show()
```

### Grouped Bar Charts

Finally, Grouped Bar Charts are a way of showing information about different sub-groups of the main categories. For a particular category separate bar charts are drawn for each of the subgroups. For example, in the following chart the results obtained for two sets of teams across a series of lab exercises are displayed. Thus each team has a bar for lab1, lab2, lab3 etc. A space is left between each category to make it easier to compare the sub categories. 


```python
#Example:

import matplotlib.pyplot as plt

width = 0.35

#set up grouped bar charts
teamA_data = [60, 75, 56, 62, 58]
teamB_data = [55, 68, 80, 73, 55] 

# Set up the index for each bar
index_teamA = [i for i in range(0, len(teamA_data))]
index_teamB = [x + width for x in index_teamA]

#Determine the mid-point for the ticks
ticks = [i + (width/2) for i in index_teamA]
tick_labels = ('Lab1', 'Lab2', 'Lab3', 'Lab4', 'Lab5')

#Plot the bar charts
plt.bar(index_teamA, teamA_data, width, color='r', label='Team A')
plt.bar(index_teamB, teamB_data, width, color='b', label='Team B')

#set up the graph
plt.xlabel('Labs')
plt.ylabel('Scores')
plt.title('Scores by Lab')
plt.xticks(ticks, tick_labels)
plt.legend()

plt.show()
```

```python
#Example 2: If we had 3 teams

width = 0.26

teamA_data = [60, 75, 56, 62, 58]
teamB_data = [55, 68, 80, 73, 55]
teamC_data = [70, 40, 70, 90, 70]

indexA = [i for i in range(0, len(teamA_data))]
indexB = [i + width for i in indexA]
indexC = [i + (2 * width) for i in indexA]

ticks= [i + width for i in range(0, len(teamA_data))]
tick_label = ['Lab1', 'Lab2', 'Lab3', 'Lab4', 'Lab5']

plt.bar(indexA, teamA_data, width, label='team A')
plt.bar(indexB, teamB_data, width, label='team B')
plt.bar(indexC, teamC_data, width, label='team C')

plt.xlabel('Labs')
plt.ylabel('Scores')
plt.title('Scores by Lab')
plt.xticks(ticks, tick_label)
plt.legend()

plt.show()
```

```python
#Example: If we had 4 teams

width = 0.2

teamA_data = [60, 75, 56, 62, 58]
teamB_data = [55, 68, 80, 73, 55]
teamC_data = [70, 40, 70, 90, 70]
teamD_data = [80, 70, 60, 95, 65]

indexA = [i for i in range(0, len(teamA_data))]
indexB = [i + width for i in indexA]
indexC = [i + (2 * width) for i in indexA]
indexD = [i + (3 * width) for i in indexA]

ticks= [i + (1.5 * width) for i in range(0, len(teamA_data))]
tick_label = ['Lab1', 'Lab2', 'Lab3', 'Lab4', 'Lab5']

plt.bar(indexA, teamA_data, width, label='team A')
plt.bar(indexB, teamB_data, width, label='team B')
plt.bar(indexC, teamC_data, width, label='team C')
plt.bar(indexD, teamD_data, width, label='team D')

plt.xlabel('Labs')
plt.ylabel('Scores')
plt.title('Scores by Lab')
plt.xticks(ticks, tick_label)
plt.legend()

plt.show()
```

A bar chart can also be a good choice for plotting histograms of bucketed numeric values, in order to visually explore how the values are distributed.
 

```python
#Example: This example shows the distriution of grades in a class.
from collections import Counter

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]

#anon func to convert the grades using Floor Division. E.g 86//10 * 10 = 80
decile = lambda grade: grade // 10 * 10

#use Counter to get a dict of converted grades and their counts. See cell below
grade_count = Counter(decile(grade) for grade in grades)

#prepare the index and heights
index = [x for x in grade_count.keys()]
grade_no = grade_count.values()

#set up chart
width = 6
plt.bar(index, grade_no, width)
plt.axis([-5, 105, 0, 5])#sets axis range

plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decile") 
plt.ylabel("No of Students") 
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

# Figures and Subplots

A Matplotlib ﬁgure is the object that contains all the graphical elements displayed on a plot. In many cases the ﬁgure is implicit as the developer interacts with the pyplot API; however the ﬁgure can be accessed directly if required. 

The matplotlib.pyplot.ﬁgure() function generates a ﬁgure object. This function returns a matplotlib.ﬁgure.Figure object. It is then possible to interact directly with the ﬁgure object. For example it is possible to add axes to the ﬁgure, to add sub plots to a graph etc. 

Working directly with the ﬁgure is necessary if you want to add multiple subplots to a ﬁgure. This can be useful if what is required is to be able to compare different views of the same data side by side. Each subplot has its own axes which can coexist within the ﬁgure. 

One or more subplots can be added to a ﬁgure using the ﬁgure.add_ subplot() method. This method adds an Axes to the ﬁgure as one of a set of one or more subplots. A subplot can be added using a 3-digit integer (or three separate integers) describing the position of the subplot. The digits represent the number of rows, columns and the index of the sub plot within the resulting matrix.

Thus 2, 2, 1 (and 221) all indicate that the subplot will take the 1st index within a two by two grid of plots. In turn 2, 2, 3 (223) indicates that the sub plot will be at index 3 which will be row 2 and column 1 within the 2 by 2 grid of plots. Where as 2, 2, 4 (or 224) indicates that the plot should be added as at index 4 or the fourth subplot within the grid (so position 2 by 2) etc. 

```python
#Example:
import matplotlib.pyplot as plt

x_values = list(range(0,20))
y_values = list(range(30, 10, -1))

#set up the grid of subplots to be 2 by 2
grid_size='22'

#initialize a figure
figure = plt.figure()

#add first subplot
position = grid_size + '1'
print('Adding first subplot to position', position)
axis1 = figure.add_subplot(position)
axis1.set(title='subplot (2,2,1)')
axis1.plot(x_values, y_values)

#add second subplot
position = grid_size + '2'
print('Adding second subplot to position', position)
axis2 = figure.add_subplot(position)
axis2.set(title='subplot (2,2,2)')
axis2.plot(x_values, y_values, 'r-')

#add third subplot
position = grid_size + '3'
print('Adding third subplot to position', position)
axis3 = figure.add_subplot(position)
axis3.set(title='subplot (2,2,3)')
axis3.plot(x_values, y_values, 'g:')

#add fourth subplot
position = grid_size + '4'
print('Adding fourth subplot to position', position)
axis4 = figure.add_subplot(position)
axis4.set(title='subplot (2,2,4)')
axis4.plot(x_values, y_values, 'y.-')

plt.show()
```
